```python

def searchInMatrix(m, s):
  y,i,j = len(m), 0, len(m[0])-1
  while i < y and j>=0:
    cell = m[i][j]
    if cell == s: return True
    elif cell > s: j -= 1
    else: i += 1

  return False

```

```python
searchInMatrix([[1,3,5,7],[10,11,16,20],[23,30,34,60]], 3)
```

```python
from collections import Counter
def mostFrequentWordSize(s):
  words = s.split()
  counter = Counter(words)
  target = counter.most_common(1)[0][0]
  print(target)
  size = len(target)

  return size
```

```python
print(mostFrequentWordSize("how can I find the most frequent string in an big string"))
print(mostFrequentWordSize("how can I find how how how string"))
```
